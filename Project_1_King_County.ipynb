```python
# Instalando a lib para descompactar as cidades em 'zipcode'

!pip install uszipcode
from uszipcode import SearchEngine
```

```python
#Criando uma nova coluna vazia para cidade
df['City'] = np.nan

cidades = SearchEngine()
for index, row in df.iterrows():
  zipcode = cidades.by_zipcode(df['zipcode'][index])
  df['City'][index] = zipcode.major_city
```

```python
# Conferindo a nova coluna com as cidades

df.head(20)
```

‚úÖ   *Encontrar a cidade de cada im√≥vel*

- Ao analizar a nova coluna percebemos que alguns im√≥veis foram vendidos mais de uma vez



---



```python
# Criado uma nova coluna, calculando o pre√ßo pela √°rea dos im√≥veis
# Chamando a nova coluna de Price_area

for index, row in df.iterrows():
  df['Price_area'] = df['price'] / df['sqft_living']
```

```python
df.head()
```

‚úÖ Calcular o pre√ßo por √°rea de im√≥veis



---



```python
# Gr√°fico mostrando o pre√ßo por √°rea dos im√≥veis em cada cidade

fig = plt.figure(figsize=(20, 6))
plt.title('Valor dos Im√≥veis de acordo com a √°rea em que est√£o localizados', loc = 'center', fontsize = 18, fontweight = 'bold')
plt.ylabel('Pre√ßo', fontsize = 18, fontweight = 'bold')
plt.grid(b=True, color='black', alpha= 0.5)
plt.rcParams['font.size'] = 12

# Gr√°fico do tipo barras 

plt.bar(df["City"], df["Price_area"], color='black')
plt.xticks(rotation=40);
```
